While performing a linear regression, we use the y=mx+c formula where y is the dependent variable and x is the independent variable.

```python
model = smf.ols(formula = 'mpg ~ horsepower' , data = df_notnull)
```

```python
result = model.fit()
```

```python
print(result.summary())
```

the below values can be used in these parameters in our formula for the line, y= (-0.1578)x+39.9359
for every unit increase in horsepower the mpg decreases by 0.1578
Intercept     39.9359     
horsepower    -0.1578

```python
# To get the coefficients alone
print(result.params)
```

```python
# To get a confidence interval
print(result.conf_int())
```

```python
#Using the linear regression
lr = linear_model.LinearRegression()
```

```python
predicted = lr.fit(X=df_notnull['horsepower'].values.reshape(-1,1),y=df_notnull['mpg'])
```

```python
print(predicted.coef_)
```
